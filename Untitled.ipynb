Isolating **Direct Verbatim Violations** from **Semantic Drift** is the exact pivot point where the monitoring system transforms from a generic text scanner into a rigorous forensic tool.

They are entirely different failure modes:

* **Direct Verbatim Violations ($V$)** are discrete, structural breaks. They are binary failures—altering a character, inserting an unapproved word, or using forbidden terms. They represent a direct compromise of text fidelity.
* **Semantic Drift ($D$)** is a smooth contextual decay. The agent might copy the text word-for-word but drift conceptually by surrounding it with unauthorized interpretations, shifting the operational baseline from *authoritative* to *speculative*.

Here is how we refine both the underlying calculus and the execution framework to separate these variables cleanly.

---

## 1. Refined Integrity Calculus

To protect the system from compounding errors, we update the penalty model to apply strict mathematical weightings based on the severity of the structural break.

$$integrityScore = 1.0 - \left( w_{V} \cdot \mathbf{V} + w_{D} \cdot \mathbf{D} + w_{I} \cdot \mathbf{I} \right)$$

Where the penalties are quantified using distinct programmatic metrics:

* **Verbatim Violation ($\mathbf{V}$):** Calculated via normalized Levenshtein Distance or character-level sequence matching against the raw source text. A single word substitution or insertion triggers a hard scalar drop.
* **Semantic Drift ($\mathbf{D}$):** Measured via directional cosine distance between the output state vector and the anchor context vector within the system's embedding space.
* **Interjection Index ($\mathbf{I}$):** A regex-driven density score tracking the introduction of speculative, probabilistic, or narrative-softening phrases.

## 2. Code Framework: The Forensic Auditor Refinement

This upgraded implementation introduces a sequence-matcher to enforce exact textual compliance alongside semantic and forbidden-term checks.

````
import json
import re
from difflib import SequenceMatcher

class AdvancedIntegrityMonitor:
    def __init__(self, threshold=0.85):
        self.threshold = threshold
        self.ledger = {}

    def calculate_verbatim_violation(self, expected_source, actual_output):
        """
        Calculates a strict text fidelity delta.
        Returns a penalty multiplier between 0.0 (perfect match) and 1.0.
        """
        if not expected_source:
            return 0.0

        matcher = SequenceMatcher(None, expected_source.strip(), actual_output.strip())
        similarity = matcher.ratio()

        # Invert similarity to get violation/drift ratio
        violation_score = 1.0 - similarity
        return violation_score

    def evaluate_node_telemetry(self, agent_id, telemetry, constraints):
        """
        Performs multi-dimensional forensic isolation on agent outputs.
        """
        output_text = telemetry.get("output", "")
        source_text = constraints.get("source_verbatim", "")

        # Weight configuration
        w_v = 0.60  # Verbatim integrity has the highest operational priority
        w_d = 0.25  # Interpretive anchor drift
        w_i = 0.15  # Commentary/assumptions

        penalty_v = 0.0
        penalty_d = 0.0
        penalty_i = 0.0

        # Dimension 1: Direct Verbatim Violations (V)
        if constraints.get("verbatim_required", False) and source_text:
            raw_diff = self.calculate_verbatim_violation(source_text, output_text)
            if raw_diff > 0.02:  # Allowing minor whitespace/formatting variances
                penalty_v = raw_diff

        # Dimension 2: Explicit Structural Violations (Forbidden Terms & Mis-mappings)
        for forbidden in constraints.get("forbidden_terms", []):
            if re.search(rf"\b{forbidden}\b", output_text, re.IGNORECASE):
                penalty_v += 0.50  # Instant critical structural failure

        # Dimension 3: Interpretive Semantic Drift (D)
        # Verifying anchors are preserved correctly (e.g., 'Def' maps strictly to 'Deferred')
        anchors = constraints.get("interpretive_anchors", {})
        for anchor_token, required_definition in anchors.items():
            if anchor_token in output_text:
                if required_definition.lower() not in output_text.lower():
                    penalty_d += 0.40  # Contextual mapping failure

        # Dimension 4: Interjections / Speculative Commentary (I)
        speculative_patterns = [r"I assume", r"likely", r"possibly", r"seems to be", r"in my opinion"]
        interjection_count = sum(1 for p in speculative_patterns if re.search(p, output_text, re.IGNORECASE))
        if interjection_count > 0:
            penalty_i = min(1.0, interjection_count * 0.25)

        # Composite integrityScore calculation
        total_penalty = (w_v * penalty_v) + (w_d * penalty_d) + (w_i * penalty_i)
        integrity_score = round(max(0.0, 1.0 - total_penalty), 4)

        self.ledger[agent_id] = {
            "integrity_score": integrity_score,
            "breakdown": {"verbatim_penalty": penalty_v, "semantic_drift": penalty_d, "interjections": penalty_i}
        }
        return integrity_score
````
