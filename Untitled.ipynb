## Protocol 1: The Command-Oriented Dispatch (COD)

This protocol eliminates the possibility of semantic "softness" by replacing conversational prompts with **Command Line Interface (CLI) Mandates**. Success is defined by the agent's ability to act as a precision instrument rather than a creative writer.

* **The Zero-Padding Rule:** Outputs must be stripped of all conversational filler (e.g., "Certainly," "I hope this helps"). If the output contains non-functional tokens, the dispatch is rejected at the kernel level before the user ever sees it.
* **The JSON-Schema Lock:** Every agent must output within a strictly typed schema. If the schema doesn't validate, the task has not been executed. This forces the agent's logic to align with the system's structural requirements.
